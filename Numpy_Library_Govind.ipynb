**Importing the Numpy Library**

```python
import numpy as np
```

```python
Arr1 = np.array([20, 30, 40, 50, 60])
Arr2 = np.array([10, 20, 30, 40, 50])

print("Arr1", Arr1)
print("Arr2", Arr2)
```

```python
# maths operation
print("Addition", Arr1 + Arr2)
print("Subtraction", Arr1 - Arr2)
print("Multiplication", Arr1 * Arr2)
print("Division", Arr1 / Arr2)
```
